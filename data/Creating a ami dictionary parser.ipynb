### Make a streaming parser

```python
import xml.sax
from xml.sax import make_parser

class dictParser(xml.sax.handler.ContentHandler):
    
    def __init__(self):
        
        #set up the counter
        self.n = 0
        
        #set up the tool we will use to store the dictionary we are extracting
        self.wordStore = []
        
        #set up the dictionary for looking elements up
        self.dictionary = {}
        
        #initialise the working dictionary
        self.working_dict = {'synonyms':[]}
        
        #initialise the flag for reading entries
        self.inEntry = False
        
        #initialise the flag for reading synonym text
        self.inSynonym = False
        
        self.descriptor = ''
    
    def startElement(self, name, attrs):
        
        if name == 'entry':
            
            #extract the items
            for key in attrs.keys():
                
                if key == 'name':

                    self.el_name = attrs[key]
                
                self.working_dict[key] = attrs[key]
            
            self.inEntry = True
            
        elif name == 'synonym':
            
            self.inSynonym = True
            
            #print('START', name)
            #print(dir(attrs))
            #print(attrs.getValue(name))

            #print('==============')
            
            
    def characters(self, content):
        
        #only want to do this if we are looking at a synonym
        if self.inSynonym:

            self.descriptor += content
        
    def endElement(self, name):
        
        if name == 'entry':
            
            #count how many entries we have
            self.n += 1
            
            #turn the entry flag off
            self.inEntry = False
            
            #store the working dictionary
            self.wordStore.append(self.working_dict)
            

            #store this work in the full dictionary
            self.store_word()
                
            #reset the working dictionary
            self.working_dict = {'synonyms':[]}
            
            #this was for debugging
            #if self.n >= 30:
            #    print(stop)
            
        elif name == 'synonym':
            
            #print('STORE SYNONYM')
            
            #store the synonym
            self.working_dict['synonyms'].append(self.descriptor)
            
            #wipe the text store object
            self.descriptor = ''
            
            #reset the flag
            self.inSynonym = False
            #print('************************')
            

    def store_word(self):
        
        #get the word in question
        name = self.working_dict['name']
        
        #add this word to the master dictionary if we don't already have it
        if self.working_dict['name'] not in self.dictionary.keys():

            self.dictionary[name] = {'synonyms':set()}
            
            #store the information for this key
            for key in self.working_dict.keys():

                if key != 'name' and key != 'synonyms':

                    self.dictionary[name][key] = self.working_dict[key]
                
                #store the synonyms
                elif key == 'synonyms':
                    
                    for syn in self.working_dict['synonyms']:
                        self.dictionary[name]['synonyms'].add(syn)
                    
                    
                    
        #if we already have an example of this word
        else:
            
            for key in self.working_dict.keys():
                
                #skip the name key
                if key == 'name':
                    continue
                    
                #check if this object exists in our storage dictionary
                if key not in self.dictionary[name].keys():
                    
                    self.dictionary[name][key] = self.working_dict[key]
                    
                #if this object has been previously stored as a single string
                elif isinstance(self.dictionary[name][key], str):
                    
                    #if we have the same word multiple times and a changing property
                    #we will save the changing value
                    if self.dictionary[name][key] != self.working_dict[key]:
                        
                        #convert the dictionary thing to a list
                        self.dictionary[name][key] = set([self.dictionary[name][key]])
                        
                        #add our new value to the list
                        self.dictionary[name][key].add(self.working_dict[key])
                        
                        
                    #if this information is already stored lets just move on
                    else:
                        continue
                        
                #if this thing is already a set lets add to it
                elif isinstance(self.dictionary[name][key], set):

                    #symptoms are lists of strings
                    if key != 'synonyms':
                    
                        if self.working_dict[key] not in self.dictionary[name][key]:

                            self.dictionary[name][key].add(self.working_dict[key])
                            
                    elif key == 'synonyms':
                        
                        for sym in self.working_dict[key]:

                            self.dictionary[name][key].add(sym)
                    
                    
        
        
###################################
### End of function definitions ###
###################################

dict_filename = './dictionaries/diseases/disease.xml'
#dict_filename = './dictionaries/diseases/disease_new.xml'
#dict_filename = './dictionaries/country/country.xml'


dictParser_tool = dictParser()

saxparser = make_parser()

saxparser.setContentHandler(dictParser_tool)

source = open(dict_filename, 'r')
        
saxparser.parse(source)

print('done')
```
